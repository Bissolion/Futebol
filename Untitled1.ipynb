```python
import pandas as pd 

df = pd.read_csv('resultado_desarmes.csv')
df.head()
```

```python
df = df[df['Minutos'] >= 900]

# Calcular a eficiência (evitando divisão por zero)
df['eficiencia'] = df['Desarmes Ganhos'] / df['Desarmes Total']
df['eficiencia'] = df['eficiencia'].fillna(0)  # Substituir valores NaN por 0 caso Desarmes Total seja zero

# Calcular a nova métrica
df['IDO'] = (
    (df['SCA DEF'] + df['GCA DEF'] + df['Desarmes(3/3)'])
    * df['eficiencia']
)

# Ordenar pela nova métrica em ordem decrescente
df = df.sort_values(by='IDO', ascending=False)
df.index = range(1, len(df) + 1)
df.head(10)
```

```python
df = df.drop('Desarmes(2/3)', axis=1)
```

```python
df.to_csv("IDO.csv")
```
